```
import json
import time
fd = open("Records.json","r")
js = fd.read()
js
```

```
record = json.loads(js)
record
```

```
print("--------------MENU--------------\n")
for key in record.keys():
    print(key,":",record[key]['Name'],"|",record[key]['Price'],"|",record[key]['Qn'])
print("\n--------------------------------\n")
ui_name = input("Enter you name           : ")
ui_mail = input("Enter you email-ID       : ")
ui_ph = input("Enter your phone number    : ")
ui_pr = str(input("Enter product id       : "))
ui_qn = int(input("Enter product quantity : "))
if(record[ui_pr]['Qn']>=ui_qn):
  print("\n--------------Bill--------------\n")
  print("Product ID     :",ui_pr)
  print("Name           :",record[ui_pr]['Name'])
  print("Price          :",record[ui_pr]['Price'],"INR")
  print("Quantity       :",ui_qn)
  print("\n--------------------------------")
  print("Billing Amount : ",record[ui_pr]['Price']*ui_qn,"INR")
  print("--------------------------------\n")
  record[ui_pr]['Qn'] = record[ui_pr]['Qn'] - ui_qn
  sales = ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]['Name']+","+str(ui_qn)+","+str(record[ui_pr]['Price']*ui_qn)+","+time.ctime()+"\n"
  fd = open("Sales.txt","a")
  fd.write(sales)
  fd.close()
else:
  print("Sorry there is a lack of products in our inventory!")
  print("We have only",record[ui_pr]['Qn'],"quantity")
  response = input("Press Y/y to purchase the available inventory...")
  if(response == "Y" or response == "y"):
    print("\n--------------Bill--------------\n")
    print("Product ID     :",ui_pr)
    print("Name           :",record[ui_pr]['Name'])
    print("Price          :",record[ui_pr]['Price'],"INR")
    print("Quantity       :",record[ui_pr]['Qn'])
    print("\n--------------------------------")
    print("Billing Amount : ",record[ui_pr]['Price']*record[ui_pr]['Qn'],"INR")
    print("--------------------------------\n")
    sales = ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]['Name']+","+str(record[ui_pr]['Qn'])+","+str(record[ui_pr]['Price']*record[ui_pr]['Qn'])+","+time.ctime()+"\n"
    record[ui_pr]['Qn'] = 0
    fd = open("Sales.txt","a")
    fd.write(sales)
    fd.close()
  else:
    print("--------------------------------\n")
    print("Thank you!")
    print("--------------------------------\n")
js = json.dumps(record)
fd = open("Records.json","w")
fd.write(js)
fd.close()
```
